```python
%%time
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
#for i,tweet in enumerate(sntwitter.TwitterSearchScraper('COVID Vaccine lang:en geocode:40.77937310313686,-117.13671577518582,900km since:2022-02-01 until:2022-03-01').get_items()):
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('COVID Vaccine lang:en since:2022-03-01 until:2022-06-01').get_items()):
    if i>5000000:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.likeCount, tweet.retweetCount, tweet.user.username])
    
# Creating a dataframe from the tweets list above
df = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'comment', 'like_count', 'retweet_count', 'Username'])
```
